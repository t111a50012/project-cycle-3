### An Advanced Multi-Dimensional Comparison of the Proportions of Sad or Hopeless Feelings Among Adolescents

### 一、 研究主題與多族群架構 (Research Topic & Multi-Group Framework)  
  本研究基於美國 CDC 的 `YRBS_2007.csv` 數據集，聚焦於「問題 3：探討青少年感到悲傷或絕望的學生比例是否存在顯著差異」。  
  為了超越基本要求的二元比較並增加統計分析的難度與廣度，本專案將反應變數（Response Variable）鎖定為二元重編碼後的 `SadOrHopeless`（悲傷或絕望感比例），並透過引入不同的分組變數（Grouping Variables），將研究對象拆解為以下**三個獨立維度的族群進行兩樣本推論比較**：
  
  1. **維度一：生理與人口統計族群（基礎任務）**
     * **比較族群**：女學生（Female） vs 男學生（Male）  
     * **研究意義**：檢定性別在青少年心理健康發展上是否存在基礎的比例差異。
  2. **維度二：校園環境與安全知覺族群（進階升級 A）**
     * **比較族群**：因安全顧慮不敢上學的學生（Safety Concerns） vs 正常上學的學生（No Concerns）  
     * **研究意義**：探討外在環境壓力（學校安全感）是否與高比例的內在悲傷感有顯著的風險相關。
  3. **維度三：物質使用與風險行為族群（進階升級 B）**
     * **比較族群**：目前有飲酒習慣的學生（Current Alcohol Users） vs 目前不飲酒的學生（Non-Alcohol Users）  
     * **研究意義**：評估成癮物質（酒精）的使用，與青少年心理負面情緒比例之間的關聯深度。

* **English Description**:  
  This study is based on the CDC's `YRBS_2007.csv` dataset, strictly focusing on **"Question 3: Examining whether there is a difference in the proportion of students who feel sad or hopeless."** To exceed default expectations and elevate the statistical complexity, our response variable is strictly locked as the binary-recoded `SadOrHopeless`. By incorporating diverse grouping variables, we partition the research into **three independent dimensional groups for two-sample inferential comparisons**:
  
  1. **Dimension 1: Physiological & Demographic Groups (Baseline Task)**
     * **Comparison**: Female Students vs. Male Students  
     * **Significance**: Testing whether biological sex presents a baseline difference in adolescent psychological well-being.
  2. **Dimension 2: School Environment & Safety Perception Groups (Advanced Upgrade A)**
     * **Comparison**: Students with Safety Concerns (missed school) vs. Students with No Concerns  
     * **Significance**: Investigating if external environmental stressors (lack of school safety) are associated with a higher proportion of internal sadness.
  3. **Dimension 3: Substance Use & Risk Behavior Groups (Advanced Upgrade B)**
     * **Comparison**: Current Alcohol Users vs. Non-Alcohol Users  
     * **Significance**: Evaluating the statistical association between addictive substance use (alcohol) and negative emotional proportions in youth.

### 二、 研究手法與資料清洗規則 (Research Methodology & Recoding Rules)
* **中文說明**：  
  本研究將嚴格遵循講義與指定的二元重編碼規則（定義 $1 = $ 成功/是，$0 = $ 失敗/否），以確保推論統計的精確性：
  * **反應變數**：`SadOrHopeless` $\rightarrow$ 代碼 1 轉為 `1`（有悲傷感）；代碼 2 轉為 `0`（無悲傷感）。
  * **性別分組**：`WhatIsYourSex` $\rightarrow$ 代碼 1 轉為 `Male` ；代碼 2 轉為 `Female`。
  * **環境分組**：`SafetyConcernsAtSchool` $\rightarrow$ 代碼 1（0天不敢上學）轉為 `0`（安全無憂）；代碼 2（1天以上因安全不敢上學）轉為 `1`（有安全顧慮）。
  * **行為分組**：`CurrentAlcoholUse` $\rightarrow$ 代碼 1（0天飲酒）轉為 `0`（無飲酒）；代碼 2（1天以上飲酒）轉為 `1`（目前有飲酒）。
  
  所有包含缺失值（NA）或不在上述特定篩選代碼內的樣本都將在步驟二中被完全剔除。

### 三、 統計分析方法與假設驗證 (Statistical Methods & Assumptions)
* **中文說明**：  
  由於本研究比較的三組對照組，其反應變數皆為**二元類別變數**，因此我們**絕對不使用** t 檢定，而是統一採用 **獨立雙比例 z 檢定（Two-Proportion z-Test）**：
  1. **描述性統計**：首先報告各組的樣本數（$n_1, n_2$）與觀測到的悲傷感比例（$\hat{p}_1, \hat{p}_2$）。
  2. **假設檢定**：設定顯著水準 $\alpha = 0.05$。虛無假設為 $H_0: p_1 = p_2$（群體間悲傷比例相等）；對立假設為 $H_a: p_1 \neq p_2$。
  3. **區間估計**：計算兩組獨立比例差異（$\hat{p}_1 - \hat{p}_2$）的 **95% 信賴區間（Confidence Interval）**，以評估實際效應量的大小。
  4. **統計假設檢查**：在檢定前，我們將確認兩樣本具備獨立性，且各組成功與失敗的期望次數均大於 10（符合常態逼近之要求）。
  5. **因果關係規避聲明**：由於本數據為觀察性橫斷面調查，所有結論僅闡述「統計相關性」，**絕不推導出任何因果關係**。

## 1.2 數據準備與變數重編碼
## Data Preparation and Recoding 

> **Methodological Protocol / 方法學協議**：
> This module programmatically binarizes core YRBS vectors and executes a strict listwise deletion protocol (`.dropna()`). By standardizing raw questionnaire codes into strictly bounded biological scales and integer types (`int`), this pipeline synchronizes the analytical baseline and exports a pristine dataset (`yrbs_cleaned.csv`) for rigorous downstream statistical inference.
> 
> 本模組對 YRBS 核心向量進行二元化轉換，並透過完全個案剔除法（Listwise Deletion）過濾缺失值與非目標雜訊。管線將問卷原始代碼標準化為具備嚴格數學邊界的二元指標與整數型態，旨在同步樣本分析基準，並導出高純淨度數據集供下游推論統計使用。

```python
# ======================================================================
# 階段 A：獨立執行變數重編碼 (Data Recoding)
# ======================================================================

# 1. 挑選本研究所需的核心分析原始欄位
target_cols = ['WhatIsYourSex', 'SadOrHopeless', 'SafetyConcernsAtSchool', 'CurrentAlcoholUse']
df_sub = df[target_cols].copy()

# 2. 重編碼反應變數 SadOrHopeless (1 -> 1: 有悲傷感, 2 -> 0: 無)
df_sub['Sad_Binary'] = df_sub['SadOrHopeless'].map({1: 1, 2: 0})

# 3. 💡【重大核心修正】：嚴格對齊 YRBS 官方定義 (1 -> Female 女性, 2 -> Male 男性)
df_sub['Sex_Label'] = df_sub['WhatIsYourSex'].map({1: 'Female', 2: 'Male'})

# 4. 執行新規則：SafetyConcernsAtSchool (代碼 1 轉 0，代碼 2 轉 1)
df_sub['Safety_Binary'] = df_sub['SafetyConcernsAtSchool'].map({1: 0, 2: 1})

# 5. 執行新規則：CurrentAlcoholUse (代碼 1 轉 0，代碼 2 轉 1)
df_sub['Alcohol_Binary'] = df_sub['CurrentAlcoholUse'].map({1: 0, 2: 1})

print("✅ 所有核心欄位已獨立完成重編碼映射（非目標代碼已暫時轉為 NaN 待下格清洗）。")
print("\n重編碼後的資料前五筆預覽：")
print(df_sub[['Sex_Label', 'Sad_Binary', 'Safety_Binary', 'Alcohol_Binary']].head())
```

```python
# ======================================================================
# 階段 B：缺失數據與雜訊值剔除及學術檔案導出 (Data Cleaning & Export)
# ======================================================================
initial_count = df_sub.shape[0]

# 1. 定義重編碼後我們必須保留的乾淨新欄位
clean_cols = ['Sad_Binary', 'Sex_Label', 'Safety_Binary', 'Alcohol_Binary']

# 2. 執行完全個案剔除法 (.dropna() 會自動過濾掉任何含有 NaN 的不完整資料列)
df_clean = df_sub[clean_cols].dropna().reset_index(drop=True)

# 3. 確保二元變數以標準學術計量整數型態（int）儲存
df_clean['Sad_Binary'] = df_clean['Sad_Binary'].astype(int)
df_clean['Safety_Binary'] = df_clean['Safety_Binary'].astype(int)
df_clean['Alcohol_Binary'] = df_clean['Alcohol_Binary'].astype(int)

# 4. 導出至 processed 資料夾
processed_csv_path = os.path.join(data_processed_dir, 'yrbs_cleaned.csv')
df_clean.to_csv(processed_csv_path, index=False)

# 5. 印出學術審報日誌，並「即時動態驗證」男女真實比例
print("============ ✅ Data Cleaning Protocol Executed ============")
print("缺失值與無效代碼已完全剔除（Listwise Deletion 已完成）。")
print(f"  - 原始基礎欄位總數: {initial_count} 筆")
print(f"  - 最終純淨有效樣本數: {df_clean.shape[0]} 筆")

print("\n 驗證導出的真實男女悲傷盛行率：")
verification = df_clean.groupby('Sex_Label')['Sad_Binary'].mean() * 100
for sex, rate in verification.items():
    print(f"  - {sex} 實際悲傷比例: {rate:.2f}%")

print(f"\n👉 乾淨數據集已成功儲存至:\n   👉 {processed_csv_path}")
```

數據準備與變數重編碼已成功落實。所有非目標代碼與缺失值均已被排除，獲得了結構純淨的有效樣本數據，為後續的敘述統計與推論統計奠定了正確的基礎。

Data preparation and binary recoding have been successfully completed. All non-target noise codes and missing observations have been eliminated, establishing a solid, error-free foundation for subsequent descriptive and inferential analysis.
